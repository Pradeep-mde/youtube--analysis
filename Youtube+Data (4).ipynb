```python
df['published_at'] = pd.to_datetime(df['published_at'], format='%Y-%m-%d %H:%M:%S')
df['published_at'].head()
```

```python
# Q6
df['published_month'] = pd.DatetimeIndex(df['published_at']).month
df['published_month'] 
```

```python
df.published_month
```

```python
# Q7
df["published_month"] = df["published_month"].replace({1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun', 7: 'July', 8: 'Aug', 
                                 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'})
df["published_month"]
```

```python
df
```

```python
# Q8

videos_per_month = df.groupby('published_month').size().reset_index(name='video_count')


each_month_published = videos_per_month.sort_values(by='video_count', ascending=False)
each_month_published
```

```python
# Exploratory Data Analysis - EDA

x_column = 'video_count'
y_column = 'published_month'

sns.barplot(x=x_column, y=y_column, data=each_month_published)

plt.title('Bar Plot of published Month on video count')
plt.xlabel('Published Month')
plt.ylabel('view count')
plt.show()
```

```python
# Q9
# Find out the unique of some variables 
unique_video_id = df['video_id'].nunique()
unique_channel_id = df['channel_id'].nunique()
unique_channel_title = df['channel_title'].nunique()

print("The unique video id: ",unique_video_id)
print('The unique channel id: ',unique_channel_id)
print('The unique channel title: ',unique_channel_title)
```

```python
# Q10
#Find the top10 channel names having the highest number of videos in the dataset and the bottom10
# having lowest number of videos.

channel_video_count = df['channel_title'].value_counts()

channel_sorted = channel_video_count.sort_values(ascending=False)

highest_video = channel_sorted.head(10)

print('The highest number of video in Top 10: ', highest_video)

lowest_video = channel_sorted.tail(10)
print('\nThe lowest number of video in Bottom 10: ',lowest_video)
```

```python
# Q11
# . Find the title of the video which has the maximum number of likes and the title of the video having
# minimum likes and write your inferences

maximum_likes_title = df.loc[df['likes'].idxmax(), 'title']
print("The maximum numbers likes title is: ",maximum_likes_title)

minimum_likes_title = df.loc[df['likes'].idxmin(), 'title']
print("The minimum numbers likes title is: ", minimum_likes_title)
```

```python
# Q13

df['dislike_ratio'] = df['dislikes']/df['view_count']

correlation = df['dislike_ratio'].corr(df['view_count'])
correlation

print('Correlation between dislike ratio and view count: ',correlation)



plt.figure(figsize=(10,6))
sns.scatterplot(x='view_count', y='dislike_ratio', data=df)
plt.title('Dislike ratio vs view count')
plt.xlabel('View count')
plt.ylabel('Dislike Ratio')
plt.show()
```
